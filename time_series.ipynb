```python
import numpy as np
x1=np.random.exponential(1,100)
print(x1[1:5])
x2=np.random.normal(1,1,100)
print(x2[1:5])
```

```python
x=np.zeros(2*len(x1))
for i in range(len(x1)):
    x[2*i]=x1[i]
    x[2*i+1]=x2[i]
```

```python
import pandas as pd
months = pd.date_range('1960-01', periods=200, freq='MS')
ts = pd.Series(x, index=months)
```

```python
# Visualization
from matplotlib import pyplot
pyplot.figure(figsize=(10.5, 5.5))
pyplot.plot(ts)
pyplot.grid()
pyplot.suptitle("Simulation stationary time series")
pyplot.xlabel('Months')
pyplot.ylabel('Values')
pyplot.show()
```

```python
import pandas as pd 
import io, requests 
import plotly.express as px 

year = ts.index.year 
month = ts.index.month 
month_name = ts.index.strftime('%b')

ts1 = ts.groupby([year, month, month_name]).mean()

px.line(
    ts1,
    x=month_name,
    y=ts1.values,
    color=year,
    width=1100,
    height=900,
    # title={title_text=='Year seasional time series', title_x==0.5}
).update_layout(xaxis_title="Month_name", yaxis_title="values",title_text='Year seasional time series', title_x=0.5)
```
