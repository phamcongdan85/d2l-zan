```python
import numpy as np
x1=np.random.exponential(1,100)
print(x1[1:5])
x2=np.random.normal(1,1,100)
print(x2[1:5])
```

```python
x=np.zeros(2*len(x1))
for i in range(len(x1)):
    x[2*i]=x1[i]
    x[2*i+1]=x2[i]
```

```python
import pandas as pd
months = pd.date_range('1960-01', periods=200, freq='MS')
ts = pd.Series(x, index=months)
```

```python
# Visualization
from matplotlib import pyplot
pyplot.figure(figsize=(10.5, 5.5))
pyplot.plot(ts)

pyplot.show()
```
