```python
import numpy as np 
import pylab 
import random 
n = 10000 #n is the number of steps(increase in the value of n increses the compelxity of graph) 
x = np.zeros(n) # x and y are arrays which store the coordinates of the position 
y = np.zeros(n) 
direction=["NORTH","SOUTH","EAST","WEST"] # Assuming the four directions of movement.
for i in range(1, n): 
    step = random.choice(direction) #Randomly choosing the direction of movement. 
    if step == "EAST": #updating the direction with respect to the direction of motion choosen.
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step == "WEST": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step == "NORTH": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
pylab.title("Random Walk 2-D")
pylab.plot(x, y) #plotting the walk.
pylab.plot(x[0], y[0], "ro", linewidth=100)

pylab.plot(x[n-1], y[n-1], "mD", linewidth=100)

pylab.show()
```

```python
import numpy as np 
import pylab 
import random 

p = 10 # probability excited

if p<0 or p>25:
        raise ValueError("probability excited must be in (0, 25)")

n = 10000 #n is the number of steps(increase in the value of n increses the compelxity of graph) 
x = np.zeros(n) # x and y are arrays which store the coordinates of the position 
y = np.zeros(n) 
direction=["NORTH","SOUTH","EAST","WEST"] # Assuming the four directions of movement.

L = [(x[0],y[0])]

for i in range(1, n): 
    if (x[i-1],y[i-1]) in L:
        step = random.choice(direction) #Randomly choosing the direction of movement.
    else:
        step = step = random.choices(direction, weights = (25, 25, 25+p, 25-p))[0] #Randomly choosing the direction of movement. 

    if step == "EAST": #updating the direction with respect to the direction of motion choosen.
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step == "WEST": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step == "NORTH": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
    
    L = L+[(x[i], y[i])]

v = x[n-1]/n
pylab.title("Ecited Random Walk 2-D, "+"v= {}".format(v))
pylab.plot(x, y) #plotting the walk.
pylab.plot(x[0], y[0], "ro", linewidth=100)

pylab.plot(x[n-1], y[n-1], "mD", linewidth=100)

pylab.show()
```
